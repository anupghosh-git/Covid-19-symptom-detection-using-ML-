```python
#import files
import pandas as pd
import numpy as np
import seaborn as sbn
data = pd.read_csv('Raw-Data.csv')
data.shape
```

```python
data.head()
```

```python
country = len(data.Country.dropna().unique())
age = len(data.Age.dropna().unique())
gender = len(data.Gender.dropna().unique())
symptoms = len(data.Symptoms.dropna().unique())
esymptoms = len(data.Experiencing_Symptoms.dropna().unique())
severity = len(data.Severity.dropna().unique())
contact = len(data.Contact.dropna().unique())

print("Total Combination Possible: ",country * age * gender * symptoms * esymptoms * severity * contact)
```

```python
#combination of columns
import itertools
columns = [data.Country.dropna().unique().tolist(),
          data.Age.dropna().unique().tolist(),
          data.Gender.dropna().unique().tolist(),
          data.Symptoms.dropna().unique().tolist(),
          data.Experiencing_Symptoms.dropna().unique().tolist(),
          data.Severity.dropna().unique().tolist(),
          data.Contact.dropna().unique().tolist()]

data_frame = pd.DataFrame(list(itertools.product(*columns)), columns=data.columns)
```

```python
data_frame.shape
```

```python
data_frame.head()
```

```python
#make columns by symptoms values
symptoms_list = data_frame['Symptoms'].str.split(',')

from collections import Counter
symptoms_counter = Counter(([a for b in symptoms_list.tolist() for a in b]))

for symptom in symptoms_counter.keys():
    data_frame[symptom] = 0
    data_frame.loc[data_frame['Symptoms'].str.contains(symptom), symptom] = 1

data_frame.head()
```

```python
#make columns by xperiencing_Symptoms values

esymptoms_list = data_frame['Experiencing_Symptoms'].str.split(',')

from collections import Counter
esymptoms_counter = Counter(([a for b in esymptoms_list.tolist() for a in b]))

for esymptom in esymptoms_counter.keys():
    data_frame[esymptom] = 0
    data_frame.loc[data_frame['Experiencing_Symptoms'].str.contains(esymptom), esymptom] = 1

data_frame.head()
```

```python
#Symptoms,Experiencing_Symptoms coluns deletion
data_frame = data_frame.drop(['Symptoms','Experiencing_Symptoms'],axis=1)
dummy = pd.get_dummies(data_frame.drop('Country',axis=1))
dummy['Country'] = data_frame['Country']
data_frame = dummy
data_frame.head()
```

```python
#final_data.to_csv('Cleaned-Data.csv', index=False, header=True)
data_frame.info()
```

```python
data_frame = data_frame.drop(['Gender_Female','Gender_Male','Gender_Transgender','Severity_Moderate','Severity_None','Severity_Severe','Contact_Dont-Know','Contact_No','Contact_Yes','Country','None_Experiencing','None_Sympton','Age_0-9','Age_10-19','Age_20-24','Age_25-59','Age_60+'],axis=1)
```

```python
data_frame.info()
```

```python

#preparing dataset by ignoring the infection '1' when the symptoms are '0' 

for counter in range(len(data_frame)):
    if data_frame.loc[counter,'Fever']  == 1 and data_frame.loc[counter,'Tiredness'] == 1  and data_frame.loc[counter,'Dry-Cough'] == 1  and data_frame.loc[counter,'Difficulty-in-Breathing'] == 1 and data_frame.loc[counter,'Sore-Throat'] == 1 and data_frame.loc[counter,'Pains'] == 1 and data_frame.loc[counter,'Nasal-Congestion'] == 1 and data_frame.loc[counter,'Runny-Nose'] == 1 and data_frame.loc[counter,'Runny-Nose'] == 1:
        data_frame.loc[counter,'Severity_Mild']  = 1
    else: 
            data_frame.loc[counter,'Severity_Mild']  = 0

```
